```python
import pandas as pd
```

```python
Data= pd.read_csv("Advertising.csv",index_col=0)
```

```python
Data
```

```python
Data.shape
```

```python
Data.isnull
```

```python
x = Data[['TV','Radio','Newspaper']]
```

```python
y= Data[['Sales']]
```

```python
! pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
```

```python
print(x_train.shape,"Traning data size for x")
print(x_test.shape,"Traning data size for x")
print(y_train.shape,"Traning data size for y")
print(y_test.shape,"Traning data size for y")
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
b0 = 2.80894491
b1 = 0.04614929
b2 = 0.18454379
b3 = 0.0038006

y = b0 + b1*44.5 + b2*39.3 + b3*45.1
```

```python
y
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
new_df = pd.DataFrame()
new_df = x_test
new_df['actual sales']= y_test
new_df['predicted sales']= y_pred
new_df
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

r2=r2_score(y_test,y_pred)
print("R-square:",r2)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("RMSE:",rmse)
```
